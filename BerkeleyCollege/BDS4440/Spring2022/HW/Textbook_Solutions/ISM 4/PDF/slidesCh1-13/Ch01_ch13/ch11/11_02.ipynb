# 11.2 Factorials
* Factorial of a positive integer **_n_** is written **_n!_** and pronounced “n factorial” 
* **_n_** · (**_n_** – 1) · (**_n_** – 2) · … · 1
* 1! is 1 
* 0! is defined to be 1
* 5! is the product 5 · 4 · 3 · 2 · 1, which is 120

### Iterative Factorial Approach
* Calculate 5! 

```python
factorial = 1
```

```python
for number in range(5, 0, -1):
    factorial *= number
```

```python
factorial
```
