```python
def unique(values):
    non_duplicates = []
    
    for value in values:
        if value not in non_duplicates:
            non_duplicates.append(value)
    
    non_duplicates.sort()
    return non_duplicates
```

```python
numbers = [11, 11, 2, 2, 7, 7, 5, 5, 3, 3]
unique(numbers)
```

```python
colors = ['red', 'red', 'orange', 'orange', 'yellow', 'yellow', 'green', 'green']
unique(colors)
```
