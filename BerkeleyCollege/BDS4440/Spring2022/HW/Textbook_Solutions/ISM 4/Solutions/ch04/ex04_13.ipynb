```python
def product(*args):
    """Calculate product of the arguments."""
    if len(args) == 0:
        return 0

    product = 1
    
    for arg in args:
        product *= arg
        
    return product
```

```python
product()
```

```python
product(1)
```

```python
product(1, 2, 3)
```

```python
product(1, 2, 3, 4, 5)
```
