# Feature Selection: Hypothesis testing of association using Chi2

```python
from scipy.stats import chi2_contingency,chi2
def chi_square_independence(data,alpha):
    chi_square_stat, p_value, dof, ex = chi2_contingency(data)
    chi_square_critical = chi2.ppf(1-alpha,dof)
    #print('chi_square_critical {}, alpha: {}'.format(round(chi_square_critical,4),round(alpha,4)))
    text = "chi_square_stat: {}, p_value: {}, dof: {}".format(round(chi_square_stat,4),round(p_value,4),dof)
    if (chi_square_critical<chi_square_stat) and (p_value<alpha):
        return "Reject Null hypothesis "+text
    else:
        return "Fail to reject "+text

use_less_features1 = []
for col in data.columns:
    if col!=target:
        df = pd.crosstab(data[col],data[target])
        if 'Fail' in chi_square_independence(df,0.01):
            use_less_features1.append(col)
use_less_features1
```

```python
data.drop(use_less_features1,axis=1,inplace=True)
data.head()
```

```python
data2 = pd.DataFrame()
for col in data.columns:
    if col!=target:
        df2 = pd.get_dummies(data[col])
        df2.drop(df2.columns[-1],axis=1,inplace=True)
        df2.columns = [col+'_'+i for i in df2.columns]
        data2 = pd.concat([data2,df2],axis=1)
data2[target]=data[target]
data = data2
data.head()
```

```python
use_less_features2 = []
for col in data.columns:
    if col!=target:
        df = pd.crosstab(data[col],data[target])
        if 'Fail' in chi_square_independence(df,0.01):
            use_less_features2.append(col)
use_less_features2
```

```python
data.drop(use_less_features2,axis=1,inplace=True)
data.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer,f1_score
f1_scorer = make_scorer(f1_score)
target='class'
X = data.drop(target,axis=1)
y = data[target].apply(lambda x: 1 if x=='p' else 0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

sgd_params = {'loss':['hinge','log','modified_huber'],'max_iter':[5,10,100],
              'penalty':['l1','l2'],'alpha':[0.001,0.01,0.1,1,10]}
rf_params = {'criterion':['gini','entropy'],'n_estimators':[10,20,len(data.columns)],'max_depth':[10,20,None]}
models = [(SGDClassifier(),sgd_params),(RandomForestClassifier(),rf_params)]

scores = {}
for m in models:
    model = m[0]
    parameters = m[1]
    gs = GridSearchCV(model, parameters,cv=5,scoring=f1_scorer)
    gs.fit(X_train,y_train)
    best_score = gs.best_score_
    best_params = gs.best_params_
    scores[str(model)] = (best_score,best_params)
```
