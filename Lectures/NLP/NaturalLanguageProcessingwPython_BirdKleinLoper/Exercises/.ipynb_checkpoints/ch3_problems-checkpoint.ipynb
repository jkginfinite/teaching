# Problem 9

* `\s*`: Matches any whitespace characters (including spaces, tabs, and line breaks), if any, before the opening parenthesis.
* `\(`: Matches the opening parenthesis.
* `[^)]*`: Matches any characters except closing parenthesis ), zero or more times, effectively matching the content inside the parentheses.
* `\)`: Matches the closing parenthesis.

```python
pattern = r'\s*\([^)]*\)'
result = re.sub(pattern, '', l)
print(result)
```

`dates_regex = r"\b\d{1,2}/\d{1,2}/\d{2}\b"`

* `'\b` asserts a word boundary, ensuring that te match starts and engs at a word boundary
* `'\d{1,2}'` matches one or two digits, representing the month and day
* `'/'` matches one forward slash character
* `\d{2}` matches exactly two digits, representing the year
* `\b` assert another word boundary ensuringng that the match ends at a word boundary.

```python
dates_regex = r"\b\d{1,2}/\d{1,2}/\d{2}\b"
print(nltk.regexp_tokenize("I was both 12/21/84",dates_regex))
```

`titles_regex = r"\b(?:[A-Z][a-z']*\b\s*)+"`

* `'\b'` asserts a word boundary, ensuring that te match starts and engs at a word boundary

* `(?: ... )` is a non-capturing group that contains the pattern for each word in the title.

* `[A-Z]` matches an upper case letter.

* `[a-z']*` matches lowercase letters or apostrophes to hanndle contractions or possessives

* `\b` asserts another word boundary, ensuring that the match ends at the end of a word.

* `\s*` matches any white spaces chaaracters the appears between words

```python
titles_regex = r"\b(?:[A-Z][a-z']*\b\s*)+"
print(nltk.regexp_tokenize("The book The Old Man And The Sea was written by Ernest Hemmingway.",titles_regex))
```

`currency_regex = r'\$\d+(?:\.\d{2})?'`

* `\` is used to escape the dollar sign `$`
* `\d+` matches one or more digits
* `(?:\.\d{2})` is an optional non-capturing group that matches the decimal part of the price (e.g. `.50`).
* `\.` matches the decimal point
* `\d{2}` matches exactly two digits (the decimal fraction part of the price.

```python
currency_regex = r'\$\d+(?:\.\d{2})?'
print(nltk.regexp_tokenize("The book costs $12.50",currency_regex))
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
url1 = "https://news.bbc.co.uk/"
url2 = "https://stackoverflow.com/questions/17721013/regular-expression-for-matching-titles-ex-book-title"

def extract(url):
    raw = urlopen(url).read()
    soup = str(BeautifulSoup(raw))
    pattern = r"<p>(.*?)</p>"
    matches = ''.join(re.findall(pattern,soup,re.DOTALL))
    l = re.sub(r"<.*?>","",matches)
    return l
```

```python
import re

text = '''
"eat"
"omelet"
"are"
"every"
"omelet"
"another"
'''

pattern = r'^(xr|yt|[aeiou]|(.*?[aeiou]))(.*)$'

def replace(match):
    if match.group(1):
        print('1',match.group(1))
        print('0',match.group(0))
        return match.group(0) + "way"
    else:
        return match.group(3) + match.group(2) + "ay"

result = re.sub(pattern, replace, text, flags=re.MULTILINE)

print(result)
```

```python
import nltk
from nltk.corpus import wordnet as wn

class IndexedText(object):
    def __init__(self,stemmer,text):
        self._text = text
        self._stemmer = stemmer
        self._index = nltk.Index((self._stem(word),i) for (i,word) in enumerate(text))
        
    def concordance(self,word,width=40):
        #key = self._stem(word)
        key = wn.synsets(word)[0].offset()
        wc = width/4
        for i in self._index[key]:
            lcontext = ' '.join(self._text[i-wc:i])
            rcontext = ' '.join(self._text[i:i+wc])
            ldisplay = '%*s'%(width,lcontext[-width:])
            rdisplay = '%-*s'%(width,rcontext[:width])
            print(ldisplay,rdisplay)
            
    def _stem(self,word):
        return self._stemmer.stem(word).lower()
text=nltk.corpus.gutenberg.raw('austen-persuasion.txt')
word = 'prize'
ps = IndexedText(nltk.PorterStemmer(),text)
ps.concordance(word)
print(ps._stem(word))
ls = IndexedText(nltk.LancasterStemmer(),text)
ls.concordance(word)
print(ls._stem(word))
```
